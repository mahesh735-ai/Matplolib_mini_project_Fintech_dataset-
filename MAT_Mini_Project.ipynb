```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("fintech_sales_dataset.csv")
df
```

```python
df.head()
```

```python
df.describe() # statistics know
```

```python
1. Sales Trend (Line Plot)

Problem:
“Analyze how sales change over time”
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['date'], df['sales'],
         color='#1f77b4',
         marker='o',
         linestyle='-',
         linewidth=2,
         label='Sales')

plt.title("Daily Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")

plt.xticks(rotation=45)
plt.grid(alpha=0.3)

plt.legend()
plt.tight_layout()

plt.show()
```

```python
2. Category-wise Sales (Bar Plot)

Problem:
“Which category generates highest revenue?”
```

```python
df.groupby('category')['sales'].sum().plot(kind='bar')  # simple for try
```

```python
category_sales = df.groupby('category')['sales'].sum()

plt.figure(figsize=(8,5))

plt.bar(category_sales.index,
        category_sales.values,
        color=['skyblue','pink','lightgreen'],
        edgecolor='black')

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.grid(axis='y', alpha=0.3)

# Add value labels  / its for data labels 
for i, val in enumerate(category_sales.values):
    plt.text(i, val + 50, str(val), ha='center')

plt.tight_layout()
plt.show()
```

```python
3. Region Share (Pie Chart)
Problem:
“Which region contributes most to business?”
```

```python
region_sales = df.groupby('region')['sales'].sum() 
```

```python
plt.figure(figsize=(6,6))

plt.pie(region_sales.values,
        labels=region_sales.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=['skyblue','pink','lightgreen','#FFB347'])

plt.title("Region-wise Sales Distribution")

plt.tight_layout()
plt.show()
```

```python
4. Profit Distribution (Histogram)

Problem:
“How profit is distributed?”
```

```python
plt.figure(figsize=(8,5)) # widht,hight

plt.hist(df['profit'],
         bins=10,
         color='#FDFD96',
         edgecolor='black',
         alpha=0.8)

plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")

plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
Insight:
Profit mostly falls in mid-range → stable business
Few extreme values → possible high/low performance
```

```python
5. Sales Outliers (Box Plot)
Problem

Are there extreme values?
```

```python
plt.figure(figsize=(6,5))

plt.boxplot(df['sales'],
            patch_artist=True,
            boxprops=dict(facecolor='#77DD77'),
            medianprops=dict(color='red'),
            flierprops=dict(marker='o', color='red'))

plt.title("Sales Outliers Detection")
plt.ylabel("Sales")

plt.grid(alpha=0.3)

plt.show()
```

```python
Insight:
Outliers present → unusual high/low sales
Important for anomaly detection
```

```python
6. Sales vs Profit (Scatter)
Problem: “Is higher sales always generating higher profit?
```

```python
plt.figure(figsize=(7,5))

plt.scatter(df['sales'], df['profit'],
            color='#1f77b4',
            s=80,
            alpha=0.7)

plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.grid()

plt.show()
```

```python
Insight:
Positive trend → more sales = more profit
Some points deviate → inefficiency
```

```python
 7. Marketing vs Sales (Scatter)
 Problem: “Does marketing increase sales?” 
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df['marketing_spend'],df['sales'],color='#FFB347',s=70,alpha=0.8)

plt.title('Marketing spends Vs Sales')
plt.xlabel('Marketing spend')
plt.ylabel('Sales')

plt.show()
```

```python
Insight:
Upward pattern → marketing effective
Weak relation → waste spending
```

```python
plt.figure(figsize=(10,5))

plt.stackplot(df['date'],
              df['sales'],
              df['profit'],
              labels=['Sales','Profit'],
              colors=['#AEC6CF','#FFB347'],
              alpha=0.8)

plt.legend()
plt.title("Business Growth Trend")
plt.xlabel("Date")
plt.ylabel("Value")

plt.xticks(rotation=45)
plt.grid()

plt.tight_layout()
plt.show()
```

```python
9. Subplot Dashboard 
Problem: “Create 4 charts in one screen (executive view)” 📊 
```

```python
plt.figure(figsize=(12,8))

# Line
plt.subplot(2,2,1)
plt.plot(df['date'], df['sales'], color='#1f77b4')
plt.title("Sales Trend")

# Bar
plt.subplot(2,2,2)
df.groupby('category')['sales'].sum().plot(kind='bar', color='#77DD77')
plt.title("Category Sales")

# Scatter
plt.subplot(2,2,3)
plt.scatter(df['sales'], df['profit'], color='#FFB347')
plt.title("Sales vs Profit")

# Histogram
plt.subplot(2,2,4)
plt.hist(df['profit'], bins=10, color='#CBAACB',edgecolor='black')
plt.title("Profit Distribution")

plt.tight_layout()
plt.show()
```

```python
Insight:
All insights in one screen
Best for management
```

```python
10. Heatmap (IMPORTANT ) 
Problem: “Find relationships between variables”
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Correlation
corr = numeric_df.corr()

plt.figure(figsize=(6,4))

# Heatmap
plt.imshow(corr, cmap='coolwarm', aspect='auto')

plt.colorbar()

#  Labels
plt.xticks(ticks=np.arange(len(corr.columns)), labels=corr.columns, rotation=45)
plt.yticks(ticks=np.arange(len(corr.columns)), labels=corr.columns)

# Annotate values
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(j, i, round(corr.iloc[i, j], 2),
                 ha='center', va='center', color='black')

plt.title("Correlation Heatmap (Matplotlib)")
plt.tight_layout()
plt.show()
```

```python
Insight:
marketing ↔ sales strong
sales ↔ profit strong
```
